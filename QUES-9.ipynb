```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 1, 5, 3]
y2 = [9, 1, 4, 2, 8]
```

```python
# scatter plot
plt.scatter(x, y1, label='Series 1', color='blue')
plt.scatter(x, y2, label='Series 2', color='red')
plt.title('Scatter Plot')
plt.legend()  
plt.show()
```

```python
# Line plot
plt.plot(x, y1, label='Series 1', color='blue')
plt.plot(x, y2, label='Series 2', color='red')
plt.title('Line Plot')
plt.legend()  
plt.show()
```

```python

# bar plot
plt.bar(x, y1, label='Series 1', color='blue')
plt.bar(x, y2, label='Series 2', color='red')
plt.title('Bar Plot')
plt.legend()
plt.show()
```
